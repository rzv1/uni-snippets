```python
from random import randint
from matplotlib.pyplot import plot, grid, show, axis

def a1(n, nr_sim=1):
    count = 0
    for _ in range(nr_sim):
        birthdays = [randint(1, 365) for _ in range(n)]
        if len(set(birthdays)) != len(birthdays):
            count += 1
    return count / nr_sim

def a2(n):
    p = 1
    for k in range(n):
        p *= (365-k)/365
    return 1 - p

def plotter(n, nr_sim=1):
    x = range(n)
    y1 = [a1(k, nr_sim) for k in range(n)]
    y2 = [a2(k) for k in range(n)]
    plot(x, y1, 'r')
    plot(x, y2, 'b')
    grid()
    show()

plotter(50, 800)
```

```python
from random import random
from math import dist

def ab2(level = 1):
    A, B, C, D, M = [0,0], [1,0], [1,1], [0,1], [0.5,0.5]
    for N in [500, 1000, 2000]:
        count = 0
        if level == 1:
            for _ in range(N):
                F = [random(), random()]
                if dist(F, M) < 0.5:
                    count += 1
                    plot(F[0], F[1], 'bo')
                else:
                    plot(F[0], F[1], 'ro')
        elif level == 2:
            for _ in range(N):
                F = [random(), random()]
                if dist(F, M) < min(dist(F, A), dist(F, B), dist(F, C), dist(F, D)):
                    count += 1
                    plot(F[0], F[1], 'bo')
                else:
                    plot(F[0], F[1], 'ro')
        elif level == 3:
            for _ in range(N):
                F = [random(), random()]
                FA, FB, FC, FD = dist(F, A), dist(F, B), dist(F, C), dist(F, D)
                obtuz = ((FA ** 2 + FB ** 2) < 1) + ((FA ** 2 + FD ** 2) < 1) + ((FB ** 2 + FC ** 2) < 1) + ((FC ** 2 + FD ** 2) < 1)
                if obtuz == 2:
                    count += 1
                    plot(F[0], F[1], 'bo')
                else:
                    plot(F[0], F[1], 'ro')
        print("Frecventa relativa cu " + str(N) + " puncte pentru cerinta " + str(level) + " este: " + str(count/N))

ab2(3)
axis('square')
show()
```
