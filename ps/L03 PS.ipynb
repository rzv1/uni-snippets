```python
from random import randint


def MontyHall_sim(schimba_usa=False, n_sim=1, afisare=False):
    count = 0
    for i in range(n_sim):
        w = randint(1, 3) #winning position
        c = randint(1, 3) #my choice
        l = randint(1, 3) #losing position
        while l == c or l == w: #reroll losing position if not correctly chosen
            l = randint(1, 3)
        c1 = c
        if schimba_usa:
            c1 = randint(1, 3)
            while c1 == l or c == c1: #reroll
                c1 = randint(1, 3)
        if c1 == w:
            count += 1
    if afisare:
        print("Probabilitatea de a castiga: " + str(count / n_sim))
    return count / n_sim

MontyHall_sim(schimba_usa=True, n_sim=10000, afisare=True)
MontyHall_sim(schimba_usa=False, n_sim=10000, afisare=True)
```
