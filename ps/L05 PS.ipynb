```python
from scipy.stats import uniform
from matplotlib.pyplot import hist, bar, show, grid, legend, plot

def ap1(nr_sim=1):
    grupe = ['0', 'A', 'B', 'AB']
    probs = [0.46, 0.40, 0.10, 0.04]
    u = uniform.rvs(size=nr_sim)
    rez = []
    for val in u:
        if val < 0.04:
            rez.append("AB")
        elif val < 0.14:
            rez.append("B")
        elif val < 0.54:
            rez.append("A")
        else:
            rez.append("0")
    #Alternativa rapida
    #rez= choice(grupe, size=1000, p=probs)
    bin_edges = [k-0.5 for k in range(5)]
    grid()
    hist(rez, bins=bin_edges, density=True, rwidth=0.8, alpha=0.5, color='r', label="Frecvente relative")
    bar(grupe, probs, width=0.5, label="Probabilistic")
    legend()

ap1(10000)
```

```python
from matplotlib.pyplot import xticks
from scipy.stats import expon
from numpy import linspace

def ap2(nr_sim=1):
    lambda_value = 1/12
    data = expon.rvs(scale=1/lambda_value, size=nr_sim)
    hist(data, bins=12, range=(0,60), density=True, rwidth=0.8, alpha=0.5, color='r', label="Frecvente relative")
    x = linspace(0, 60, 200)
    pdf = expon.pdf(x, scale=1/lambda_value)
    plot(x, pdf, 'r-')
    xticks(range(0, 61, 5))
    grid()
    show()
    print("Probabilitate eveniment E este " +
          str(1-expon.cdf(5, scale=1/lambda_value)))

    print("Probabilitate eveniment E din simulari este " +
          str(sum([x>=5 for x in data])/len(data)))

ap2(10000)
```
