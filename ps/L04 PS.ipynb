```python
from scipy.stats import bernoulli, binom
from matplotlib.pyplot import bar, hist, show, legend, xticks, grid

def a1(n, p, afisare=False):
    pos = [0]
    for _ in range(n):
        x = bernoulli.rvs(p)
        pos.append(pos[-1] + 1 if x == 1 else pos[-1] - 1)
    if afisare:
        print(pos)
    return pos[-1]

def b1(n, p):
    data = [a1(n, p) for _ in range(1000)]
    bins = [k-0.5 for k in range(-25, 25)]
    grid()
    hist(data, bins=bins, label="Frecvente relative")
    legend()

def c1(n, p):
    data = [a1(n, p) % n for _ in range(1000)]
    bins = [k-0.5 for k in range(n)]
    grid()
    hist(data, bins=bins, label="reprezentare pe cerc")
    legend()

a1(50, 0.5, afisare=True)
b1(50, 0.5)
```
