```python
from random import sample
from math import factorial
from itertools import permutations
```

Afisati o lista cu toate permutarile cuvantului word
- permutations(iterable, r=None)

```python
l = []
for e in permutations("word"):
    l.append(e)
print(l)
```

Afisati numarul total al permutarilor cuvantului word

```python
print(len(l))
```

Afisati o permutare aleatoare
- sample(population, k, *, counts=None)

```python
print(sample(l, 1))
```

Aceiasi cerinta cu aranjamente si combinari

```python
from math import perm, comb #calculeaza numarul matematic
from itertools import permutations, combinations #le genereaza

l = []
for e in permutations("word", 2):
    l.append(e)
print(l); print(len(l)); print(sample(l, 1))

l = []
for e in combinations("word", 2):
    l.append(e)
print(l); print(len(l)); print(sample(l, 1))
```

Afisati toate combinarile cu repetitie de 4 elemente din grupul de litere ABCDE

```python
from itertools import combinations_with_replacement

l = []
for e in combinations_with_replacement("word", 3):
    l.append(e)
print(l);
```
